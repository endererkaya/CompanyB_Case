```python
cc_column_names = np.array(df_credit_card_balance.columns)
cc_column_names = np.delete(cc_column_names,-3)
print(df_credit_card_balance.columns[1:])
```

```python
unique_ccb_ids = df_credit_card_balance["SK_ID_CURR"].unique()

df_ccb_unique = pd.DataFrame(columns=(df_credit_card_balance.columns[1:]))
temp_index = 0
for unique_id in unique_ccb_ids:
    if temp_index%100==0:
        print(temp_index)
    temp_df = df_credit_card_balance[df_credit_card_balance["SK_ID_CURR"]==unique_id]
    temp_status = temp_df["NAME_CONTRACT_STATUS"].mode()
    temp_df_amt1 = []
    for column_name1 in cc_column_names[2:5]:
        temp_df_amt1.append(temp_df[column_name1].mean(axis=0))
    temp_df_amt2 = []
    for column_name2 in cc_column_names[5:15]:
        new_temp = temp_df[column_name2]
        temp_df_amt2.append(new_temp[new_temp>0].mean(axis=0))
    temp_df_amt3 = []
    for column_name3 in cc_column_names[15:19]:
        temp_df_amt3.append(temp_df[column_name3].sum())
    temp_df_amt4 = []
    temp_cum     = temp_df[cc_column_names[19]].mean(axis=0)
    for column_name4 in cc_column_names[20:22]:
        temp_df_amt4.append(temp_df[column_name4].sum())
    
    temp_all = [unique_id] + temp_df_amt1 + temp_df_amt2 + temp_df_amt3 + [temp_cum] + [temp_status[0]] + temp_df_amt4
    # append rows to an empty DataFrame
    df_ccb_unique.loc[temp_index,df_credit_card_balance.columns[1:]] = temp_all
    temp_index = temp_index + 1
```
