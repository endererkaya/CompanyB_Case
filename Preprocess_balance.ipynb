```python
## Import
import pandas as pd
from pandas.plotting import autocorrelation_plot
import numpy as np
import matplotlib.pyplot as plt
import math
import copy
from scipy import stats
```

```python
import os
main_data_path = r"C:\Users\erkay\B2Metric_Case\home-credit-default-risk"
bureau_dataset_path = os.path.join(main_data_path,"bureau.csv")
df_bureau           = pd.read_csv(bureau_dataset_path)
```

```python
column_names = df_bureau.columns[4:]
column_names = np.delete(column_names,-3)
```

```python
columns = (['SK_ID_CURR','NUM_CREDIT_ACTIVE','NUM_CREDIT_CLOSED','CREDIT_CURRENCY','CREDIT_TYPE']+ column_names.to_list())
print(len(columns))
```

```python
unique_bureau_ids = df_bureau["SK_ID_CURR"].unique()

df_balance_unique = pd.DataFrame(columns=(['SK_ID_CURR','NUM_CREDIT_ACTIVE','NUM_CREDIT_CLOSED','CREDIT_CURRENCY','CREDIT_TYPE']+ column_names.to_list()))
temp_index = 0
for unique_id in unique_bureau_ids:
    if temp_index%100==0:
        print(temp_index)
    temp_df = df_bureau[df_bureau["SK_ID_CURR"]==unique_id]
    temp_active = (temp_df["CREDIT_ACTIVE"]=="Active").sum()
    temp_closed = (temp_df["CREDIT_ACTIVE"]=="Closed").sum()
    temp_currency = temp_df["CREDIT_CURRENCY"].mode()
    temp_credit_type = temp_df["CREDIT_TYPE"].mode()
    temp_df_amt   = temp_df[column_names].mean(axis=0)
    # append rows to an empty DataFrame
    df_balance_unique.loc[temp_index,"SK_ID_CURR"] = unique_id
    df_balance_unique.loc[temp_index,"NUM_CREDIT_ACTIVE"] = temp_active
    df_balance_unique.loc[temp_index,"NUM_CREDIT_CLOSED"] = temp_closed
    df_balance_unique.loc[temp_index,"CREDIT_CURRENCY"] = temp_currency[0]
    df_balance_unique.loc[temp_index,"CREDIT_TYPE"] = temp_credit_type[0]
    df_balance_unique.loc[temp_index,column_names]  = temp_df_amt
    temp_index = temp_index + 1
```
